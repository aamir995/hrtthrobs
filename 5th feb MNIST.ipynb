```python
import tensorflow as tf
```

```python
# Data 
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
x_train.shape, y_train.shape
```

```python
# Data Processing
# Pixel Value can be 0 - 255
import numpy as np 
alist = np.array([5, 10, 15])
alist / alist.max()
```

```python
# data pre-processing 
# Normalize data 
x_train = x_train / 255.0
```

```python
x_train[1]
```

```python
# Modelling 
model = tf.keras.models.Sequential()
```

```python
# Input Layer
model.add(tf.keras.layers.Input(shape=(28, 28), name="inpLayer"))
model.add(tf.keras.layers.Flatten(name='flatLayer'))
```

```python
# Hidden Layer
model.add(tf.keras.layers.Dense(units = 120, activation=tf.keras.activations.relu, name="hidLayer"))
```

```python
# Output Layer 
model.add(tf.keras.layers.Dense(units = 10, activation = tf.keras.activations.softmax, name = "opLayer"))
```

```python
model.summary()
```

```python
# Param - Weights & baises .. 
# Weights - 
(784 * 120)
# Weights & Biases
(784 * 120) + 120
```

```python
(120 * 10 ) + 10
```

```python
# Feed Forward 
# Error Cal
# Back Pro
model.compile(loss=tf.keras.losses.categorical_crossentropy, 
              optimizer=tf.keras.optimizers.SGD(learning_rate=0.01), 
              metrics=["acc"])
```
