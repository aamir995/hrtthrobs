```python
# 1. Make a sequential model 
model = tf.keras.models.Sequential()
```

```python
# Input layer
model.add(tf.keras.layers.Input((None,28, 28))) # TBD 
model.add(tf.keras.layers.Flatten())
```

```python
# add hidden layer 
model.add(tf.keras.layers.Dense(120, activation=tf.keras.activations.relu)) # number of neurons, How did we decide this number ? 
```

```python
# output layer
model.add(tf.keras.layers.Dense(10, activation=tf.keras.activations.softmax))
# SoftMax will be the activation in LAST LAYER for classification task! 
```

```python
# del model
model.summary()
```

```python
import numpy as np
import math
```

```python
def softmax(alist):
  blist = []
  den = sum([(math.e**item)for item in alist])
  for item in alist: 
    blist.append(math.e**item/den)
  return blist
```

```python
result = softmax([10,30,40,45,30])
```

```python
print(result)
print(sum(result))
```

```python
# Position of max numbers! 
np.argmax(result) 
```
