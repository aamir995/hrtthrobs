7. Write a function which finds all Pythagorean triplets of triangles whose sides are no greater than a natural number N.

```python
def pythagoreanTriplets(limits) :
    c, m = 0, 2
 
    # Limiting c would limit
    # all a, b and c
    while c < limits :
         
        # Now loop on n from 1 to m-1
        for n in range(1, m) :
            a = m * m - n * n
            b = 2 * m * n
            c = m * m + n * n
 
            # if c is greater than
            # limit then break it
            if c > limits :
                break
            print(a, b, c)
 
        m = m + 1
 
 
# Driver Code
if __name__ == '__main__' :
     
    N = 40
    pythagoreanTriplets(N)
 
```

```python
#9.There are 3 arrangements of the word DAD, namely DAD, ADD, and DDA. How many arrangements are there of the word ENDURINGLY?

s = 'ENDURINGLY'
dup_num={}

for i in range(0,len(s)):
    count = 1

    for j in range(i+1,len(s)):
        if (s[i] == s[j] and s[i] != ' ' and s[i] !='0'):
            count =count + 1
            s=s[:j] +'0' + s[j+1:]

        if(count>1):
            dup_count={s[i]:count}
            dup_num.update(dup_count)

denom = 1
for i in list(dup_num.values()):
    denom = denom * math.factorial(i)

tot_arr = int(math.factorial(len(s))/ denom)

print("The number of different arrangements are:", tot_arr)
```

```python
#14. Consider the data X = (58,59,63,60,60,63,60,57,58,59). An unbiased estimation for population variance would be ____

import statistics as st
from random import randint

X = (58,59,63,60,60,63,60,57,58,59)

def myVariance(X):
    # finds the variance of a given set of numbers
    xbar = st.mean(X)
    return sum([(x - xbar)**2 for x in X])/len(X)

def myUnbiasedVariance(X):
    # finds the 'unbiased' variance of a given set of numbers (divides by N-1) 
    xbar = st.mean(X)
    return sum([(x - xbar)**2 for x in X])/(len(X)-1)

population = [randint(0, 1000) for i in range(0,100)]

print (myVariance(population))

print (myUnbiasedVariance(population))

print (st.variance(population))
```
