```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.rand()
```

```python
# INPUT; OUTPUT
# y = 3x + 5
# x = 3; y=14
# x = 2; y=11
# y = f(x) - What is the relationship between x & y

x = 2
y = 11 
w = np.random.rand() # random Number 
b = np.random.rand() # random Number
# w = np.random.randint(0, 40, 1)
# b = np.random.randint(0, 40, 1)

errorList = []
wList = []
bList = []
ypredList = []

alpha = 0.01 # learning rate! 
epoch = 50 
for i in range(epoch): # epoch
  # Feed Forward 
  y_pred = w*x + b 
  ypredList.append(y_pred)

  # Error 
  error = (y - y_pred)**2
  errorList.append(error)
  dEdw = -2*x*(y - y_pred)
  dEdb = -2*(y - y_pred)

  # Back Propogation
  w = w - alpha*dEdw
  b = b - alpha*dEdb
  wList.append(w)
  bList.append(b)
```

```python
print(f"Error {error} & last y_pred is {y_pred}")
plt.figure(figsize=(10, 8))
plt.subplot(2,2,1)
plt.plot(errorList)
plt.title("Error Graph")

plt.subplot(2,2,2)
plt.plot(ypredList)
plt.title("y_pred Value")

plt.subplot(2,2,3)
plt.plot(wList)
plt.title("w value")

plt.subplot(2,2,4)
plt.plot(bList)
plt.title("b Value")
```
