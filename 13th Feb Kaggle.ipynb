```python
# it is different from MNIST 
# we get an array with images pre-loaded 

# We have to create an array with images stacked toghher ... with shape (25000, x, x, 3)

# 1. Read the images 
# 2. BGR - RGB 
# 3. Resize 
# 4. Append in an array
# 5. We have to note the labels also (from filename)
```

```python
folder = "train/"
images = []
labels = []
for _imageName in os.listdir("train"):
  imageName = folder + _imageName
  image = cv2.imread(imageName)
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  image = cv2.resize(image, (150, 150))
  images.append(image)
  if _imageName.lower().startswith("cat"):
    labels.append(0)
  else:
    labels.append(1)
```
