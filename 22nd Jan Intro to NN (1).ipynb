```python
# Feed Forward 
# Error Cal
# Back Propogation 


# FF
# y = wx+b

# Error
# e = (y-yp)^2 = (y - wx - b)^2

# BP
# w_new = w_old - (dE/dw) 
# b_new = b_old - (dE/db)

# dE/dw = 2(y-yp)^(2-1)(0-x-0) = 2(y-yp)(-x) = -2x(y-yp)
# db/dw = 2(y-yp)^(2-1)(0-0-1) = 2(y-yp)(-1) =-2(y-yp)
```

```python
# 1/1+e^(-x)
import math
def sigmoid(x):
  num = 1 
  den = 1 + math.e**(-x)
  return num/den 
```
